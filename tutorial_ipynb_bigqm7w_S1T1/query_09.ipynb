```python
import pymoldis

df=pymoldis.get_data('bigqm7w_S1T1')
```

```python
pymoldis.print_MolFormula(df)  # to print all molecular formulas in the dataset
```

```python
Formula='C3_H4_O4'  # pick a molecular formula
indices=pymoldis.get_ConstitutionalIsomers(df,Formula)  # get the indices of the constitutional isomers for the formula
indices
```

```python
print(df[['SMI','S1_SCSPBEQIDH(eV)','S1_ADC2(eV)']].iloc[indices]) # get the properties of interest
```

```python
S1T1_DFT=df['S1_SCSPBEQIDH(eV)'] - df['T1_SCSPBEQIDH(eV)']
print(S1T1_DFT.iloc[indices])
```
