```python
import pymoldis
import matplotlib.pyplot as plt

df=pymoldis.get_data('bigqm7w_S1T1')
```

```python
s1=df['S1_ADC2(eV)']
t1=df['T1_ADC2(eV)']
stg_adc2=s1-t1  
```

```python
s1=df['S1_SCSPBEQIDH(eV)']
t1=df['T1_SCSPBEQIDH(eV)']
stg_dft=s1-t1
```

```python
plt.scatter(stg_adc2, stg_dft)
plt.xlabel('S$_1$-T$_1$ from ADC(2) (eV)')
plt.ylabel('S$_1$-T$_1$ from SCS-PBE-QIDH (eV)')
plt.title('Scatter plot')
plt.show()
```
